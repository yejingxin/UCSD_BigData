```python
# http://matplotlib.org/basemap/users/merc.html

from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
# llcrnrlat,llcrnrlon,urcrnrlat,urcrnrlon
# are the lat/lon values of the lower left and upper right corners
# of the map.
# lat_ts is the latitude of true scale.
# resolution = 'i' means use intermediate resolution coastlines.
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines(linewidth=0.5)#,color='grey')
#m.fillcontinents(color='coral',lake_color='aqua')

# draw parallels and meridians.
parallels = np.arange(-80,81,10.)
# labels = [left,right,top,bottom]
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])
lats= D2C_eff.loc[:,'latitude'].values
lons= D2C_eff.loc[:,'longitude'].values
x, y = m(lons,lats)
z= D2C_eff.loc[:,'Dist2Coast'].values
m.scatter(x, y, c=z, s=20, cmap=plt.cm.jet, edgecolors='None', alpha=0.75)
cbar = plt.colorbar()
#plt.clim(-16,32)
cbar.ax.set_ylabel('Distance to Coast')
plt.title('weather stations')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
x= D2C_eff.loc[:,'Dist2Coast'].values
y = D2C_eff.loc[:,'Eff_year'].values
 
# Estimate the 2D histogram
nbins = 1000
H, xedges, yedges = np.histogram2d(x,y,bins=nbins)
 
# H needs to be rotated and flipped
H = np.rot90(H)
H = np.flipud(H)
 
# Mask zeros
Hmasked = np.ma.masked_where(H==0,H) # Mask pixels with a value of zero
 
# Plot 2D histogram using pcolor
fig2 = plt.figure(figsize=(15,10),dpi=300)
plt.pcolormesh(xedges,yedges,Hmasked)
plt.xlabel('x')
plt.ylabel('y')
cbar = plt.colorbar()
cbar.ax.set_ylabel('Counts')
```

```python
coast_dist = D2C_eff.loc[:,'Dist2Coast'].values
coast_tile = np.zeros(5)
for i in range(3):
    coast_tile[i+1]=np.percentile(coast_dist,25*(i+1))
coast_tile[0]=np.min(coast_dist)
coast_tile[4]=np.max(coast_dist)
print coast_tile
eff_year = D2C_eff.loc[:,'Eff_year'].values
year_tile = np.zeros([4,5])
for j in range(4):
    tmp = [eff_year[e] for e in range(len(coast_dist)) if (coast_dist[e]>=coast_tile[j]) and (coast_dist[e]<coast_tile[j+1])]
    for i in range(3):
        year_tile[j,i+1]=np.percentile(tmp,25*(i+1))
    year_tile[j,0] = np.min(tmp)
    year_tile[j,4] = np.max(tmp)
print year_tile
```

```python
rows  = shape(eff_year)[0]
partition_labels = np.zeros(rows)
for k in range(rows):
    for i in range(4):
        for j in range(4):
            if coast_dist[k]>=coast_tile[i] and coast_dist[k]<coast_tile[i+1] and eff_year[k]>=year_tile[i,j] and eff_year[k]<year_tile[i,j+1]:
                partition_labels[k]=j+i*4
D2C_eff['partition_labels']= partition_labels
D2C_eff.head()
```
