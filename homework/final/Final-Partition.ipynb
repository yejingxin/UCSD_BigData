#### regenerate array according to its weight

Since variable is weighted by its number of measurement, the simple idea is repeat variable once every daily data point. Here variables are repeated every 50 data points to relief computation complexity.

```python
def weight_array(ar, weights):
     zipped = zip(ar, weights)
     weighted = []
     for i in zipped:
         for j in np.arange(0,i[1],50):
             weighted.append(i[0])
     return weighted
coast_dist = D2C_eff.loc[:,'Dist2Coast'].values
Ndays = D2C_eff.loc[:,'Ndays'].values
weighted = weight_array(coast_dist, Ndays)
```

#### Since I partition data according to their distances to coast, it is 1-D data, simple percentile function in numpy is used here.

```python
print shape(weighted)
partition = np.zeros(128+1)
qq = np.arange(0,100.02,100.0/128)
```

```python
partition = np.percentile(weighted, list(qq))
print partition
```

####Label data with their partition_labels for the use of next notebook

```python
rows  = shape(D2C_eff)[0]
Nlabels = 128
partition_labels = np.zeros(rows)
for k in range(rows):
    for i in range(Nlabels):
        if coast_dist[k]>=partition[i] and coast_dist[k]<partition[i+1]:
                partition_labels[k]=i
D2C_eff['partition_labels']= partition_labels
D2C_eff.head(20)
```

#### verify the uniformity of the partition

```python
Nmeas = np.zeros(128)
for i in range(128):
    test = D2C_eff[D2C_eff['partition_labels']==i]
    Nmeas[i]=sum(test.loc[:,'Ndays'].values)/365
plot(Nmeas)
```
